```python

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

#querystring = {"ingr":"champagne","category[0]":"generic-foods", "limit": 10}

querystring = {
  "ingr": "champagne",
  "parsed":[{"food": {"foodId": "",
                      "label": "",
                      "category": "",
                      "foodContentsLabel": "",
                      "image": ""}}]}

headers = {
	"X-RapidAPI-Key": "2a57a56e6bmsh7d4fed488c48c7ap175bcejsn6baabdc408cd",
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

output = response.json()

print(type(output))
print(output)
```

```python
l_food_Id = []
l_label = []
l_category = []
l_foodcontentslabel = []
l_image = []



def get_info(searched_label_1, searched_label_2, searched_label_3, searched_label_4, searched_label_5):
  for dico in list_1:
    #print("dico: {}".format(dico))

    if searched_label_1 in dico['food']:
      value = dico['food'][searched_label_1]
      l_food_Id.append(value)

    else:
      l_food_Id.append('no_value')

    if searched_label_2 in dico['food']:
      value = dico['food'][searched_label_2]
      l_label.append(value)

    else:
      l_label.append('no_value')


    if searched_label_3 in dico['food']:
      value = dico['food'][searched_label_3]
      l_category.append(value)

    else:
      l_category.append('no_value')


    if searched_label_4 in dico['food']:
      value = dico['food'][searched_label_4]
      l_foodcontentslabel.append(value)

    else:
      l_foodcontentslabel.append('no_value')


    if searched_label_5 in dico['food']:
      value = dico['food'][searched_label_5]
      l_image.append(value)

    else:
      l_image.append('no_value')

  print(l_image)


  d_glob = {'foodId': l_food_Id, 'label': l_label, 'category': l_category,
            'foodContentsLabel': l_foodcontentslabel, 'image': l_image}

  df_glob = pd.DataFrame(d_glob)

  df_glob = df_glob.head(10)

  return df_glob

```

```python
df_glob = get_info('foodId', 'label', 'category', 'foodContentsLabel', 'image')
```
