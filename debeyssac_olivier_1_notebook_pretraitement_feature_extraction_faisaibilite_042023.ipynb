```python
def nltk_tokenize(df):

    print('Lancement tokenization du corpus...')

    col_name = 'product_name'
    new_col_name = 'tokens'


    # Tokenize words
    df[new_col_name] = df[col_name].apply(lambda document: word_tokenize(document))
    words = df['tokens'].tolist()
    words = [word for document in words for word in document]

    c_s_1 = len(words)
    u_w_1 = len(set(words))

    # Display corpus word counts and unique words count
    print("Nombre mots corpus: {}, nombre mots uniques: {}".format(c_s_1, u_w_1))

    corpus_size = [c_s_1]
    unique_w_size = [u_w_1]


    # Lower
    s = df[col_name]
    s = s.apply(lambda liste: liste.lower())
    df['lowered_tokens'] = s

    # Exclude low frequent words
    words = df['lowered_tokens'].to_list()
    words = [word for string in words for word in string.split()]
    #print(len(words))

    rare_words = pd.Series(words)
    rare_words = rare_words.value_counts()
    rare_words = rare_words.sort_values(ascending = False)

    list_rare_words = list(rare_words[rare_words==1].index)
    #print(len(list_rare_words))

    s = df['lowered_tokens']
    s = s.apply(lambda liste: [word for word in liste.split() if word not in list_rare_words])

    df['lowered_tokens'] = s
    #print(df['lowered_tokens'][0:2])

    # Display corpus word counts and unique words count
    words = df['lowered_tokens'].tolist()
    words = [word for document in words for word in document]

    c_s_2 = len(words)
    u_w_2 = len(set(words))


#     print("Nombre mots corpus: {}, nombre mots uniques: {}".format(c_s_2, u_w_2))
    corpus_size = [c_s_1, c_s_2]
    unique_w_size = [u_w_1, u_w_2]



    # Nettoyage caractères spéciaux
    spec = ['@','*', '#', '$', '%', '+', '-', '/', '(', ')', '&']
    words = df['lowered_tokens'].tolist()
    words = [word for document in words for word in document]
    #print("avant: {}".format(words))

    s = df['lowered_tokens']
    s = s.apply(lambda liste: [word for word in liste if word not in spec])
    df['lowered_tokens'] = s

    # Display corpus word counts and unique words count
    words = df['lowered_tokens'].tolist()
    words = [word for document in words for word in document]
    #print("après: {}".format(words))
    c_s_3 = len(words)
    u_w_3 = len(set(words))

    #print("Nombre mots corpus: {}, nombre mots uniques: {}".format(c_s_3, u_w_3))
    corpus_size = [c_s_1, c_s_2,c_s_3]
    unique_w_size = [u_w_1, u_w_2,u_w_3]



    # Nettoyage string contenant chiffres et lettres.
    s = df['lowered_tokens']
    s = s.apply(lambda liste: ' '.join(w for w in liste if not any(x.isdigit() for x in w)))
    df['lowered_tokens'] = s


    # Display corpus word counts and unique words count
    words = df['lowered_tokens'].tolist()
    #print(words)
    words = [word for document in words for word in document.split()]
    # print("après: {}".format(words))
    c_s_4 = len(words)
    u_w_4 = len(set(words))


    # print("Nombre mots corpus: {}, nombre mots uniques: {}".format(c_s_4, u_w_4))
    corpus_size = [c_s_1, c_s_2,c_s_3, c_s_4]
    unique_w_size = [u_w_1, u_w_2,u_w_3, u_w_4]



    # Nettoyage len(tokens) < 3.
    min_len = 3
    s = df['lowered_tokens']
    s = s.apply(lambda doc: [word for word in doc.split() if len(word) > 3])
    df['lowered_tokens'] = s
    #print(df['lowered_tokens'][0:2])


    # Display corpus word counts and unique words count
    words = df['lowered_tokens'].tolist()
    words = [word for document in words for word in document]
    # print("après: {}".format(words))
    c_s_5 = len(words)
    u_w_5 = len(set(words))

    # print("Nombre mots corpus: {}, nombre mots uniques: {}".format(c_s_5, u_w_5))
    corpus_size = [c_s_1, c_s_2,c_s_3, c_s_4, c_s_5]
    unique_w_size = [u_w_1, u_w_2,u_w_3, u_w_4, u_w_5]

    # Enlever les stop words
    stop = stopwords.words('english')
    s = df['lowered_tokens']
    s = s.apply(lambda doc: [word for word in doc if word not in stop])
    df['lowered_tokens'] = s

    # Display corpus word counts and unique words count
    words = df['lowered_tokens'].tolist()
    # print(words)
    words = [word for document in words for word in document]

    c_s_6 = len(words)
    u_w_6 = len(set(words))

    print()
    print('Fin de tokenization du corpus.')
    print()
    print("Nombre mots corpus en fin de tokenization: {}, nombre mots uniques: {}".format(c_s_6, u_w_6))
    corpus_size = [c_s_1, c_s_2,c_s_3, c_s_4, c_s_5, c_s_6]
    unique_w_size = [u_w_1, u_w_2,u_w_3, u_w_4, u_w_5, u_w_6]

    d = {'corpus_size': corpus_size, 'unique_w_size': unique_w_size}

    df_ratio = pd.DataFrame(d)
    df_ratio['pc_corpus_size'] = df_ratio['corpus_size']/df_ratio['corpus_size'][0]
    df_ratio['pc_unique_w_size'] = df_ratio['unique_w_size']/df_ratio['unique_w_size'][0]


    # Graphe évolution taille corpus et mots uniques
    data = df_ratio.iloc[:, -2:]
    sns.lineplot(data = data)
    plt.title('Evolution taille corpus et mots uniques')
    plt.show()




    return df, df_ratio
```

### Analyse fréquence d'occurence, feature 'product_name'.

```python
def all_words(df):

    df = nltk_tokenize(df)[0]

    # Select column
    s= df['lowered_tokens']

    element = []

    for i in range(len(s)):
        element.append(s[i])

    tok_words = [word for string in element for word in string]

    return tok_words

```

```python
def freq_tok_w():
    tok_words = all_words(df)

    freq = FreqDist(tok_words)

    return freq
```

```python
# On décide que les mots dont la fréquence est > à 100 feront partie de la liste des stop words.

def update_stop_words():

    freq = freq_tok_w()

    df_freq = pd.DataFrame(freq.items(), columns = ['items', 'nb_occ'])
    df_freq = df_freq.sort_values(by='nb_occ', ascending = False)
    maxi = df_freq['nb_occ'].max()

    # On fixe comme condition que les mots dont le nb d'occurrences est > à 100 seront
    # considérés comme stop words
    nb_occ_max = 100

    cond = (df_freq['nb_occ'] >= nb_occ_max)
    data = df_freq.loc[cond, :]

    # Graphe de fréquence des mots > 100.
    sns.barplot(data = data, x = 'items', y = 'nb_occ')
    plt.xticks(fontsize = 8, rotation = 45)
    plt.show()



    # Exclure les stop words de l'analyse.
    add_stop_w = data['items'].tolist()
    #print(add_stop_w)
    #print("tok_words: {}".format(len(tok_words)))


    return add_stop_w



```

```python
def exclude_stop_words():

    add_stop_w = update_stop_words()

    #print("add_words: {}".format(add_stop_w))

    # col_name: nom colonne df de laquelle on veut exclure les stop words.
    s = df['lowered_tokens']
    s = s.apply(lambda sentence: [word for word in sentence if word not in add_stop_w])

    df['lowered_tokens'] = s

    return df





```

## Nettoyage texte, spacy.

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
def spacy_tokenize(df, remove_stopwords=False, n_process = 1):

    texts = df['product_name']

    docs = nlp.pipe(texts,
                   n_process=n_process,
                   disable=['parser', 'ner',
                           'lemmatizer', 'textcat'])

    stopwords = nlp.Defaults.stop_words

    docs_cleaned = []
    for doc in docs:
        tokens = [tok.text.lower().strip() for tok in doc if not tok.is_punct]


        if remove_stopwords:
            tokens = [tok for tok in tokens if tok not in stopwords]

        doc_clean = ' '.join(tokens)
        docs_cleaned.append(doc_clean)


    return docs_cleaned

```

```python
def text_processing_execution():

    choix = str(input("entrer 'nltk' ou 'spacy' comme choix de méthode"))
    print("Vous avez entré: {}".format(choix))

    if choix == 'nltk':
        exclude_stop_words()

        # Display corpus word counts and unique words count
        words = df['lowered_tokens'].tolist()
        words = [word for string in words for word in string]


    else:
        df['lowered_tokens'] = spacy_tokenize(df, remove_stopwords=False, n_process = 1)


        # Display corpus word counts and unique words count
        words = df['lowered_tokens'].tolist()
        words = [word for string in words for word in string.split()]


    c_s = len(words)
    u_w = len(set(words))
    print("Nombre mots corpus: {}, nombre mots uniques: {}".format(c_s, u_w))


    return df

```

```python
def stem(df, col_name):

    s = df[col_name]

    trans = PorterStemmer()

    s = s.apply(lambda doc: [trans.stem(word) for word in doc])

    df[col_name] = s

    return df
```

```python
def lem(df, col_name):

    s = df[col_name]

    trans = WordNetLemmatizer()

    s = s.apply(lambda doc: [trans.lemmatize(word) for word in doc])

    df[col_name] = s

    return df
```

### Constitution des clusters.

A la suite de l'applicaiton de la méthode TF-IDF, on a maintenant un ensemble de most qui ont leur représentation dans un espace vectoriel. Ces vecteurs peuvent être plus ou moins proches les uns des autres selon leur qu'ils caractérisent le document ou non. En faisant tourner un kmeans sur cet espace vectoriel, on devait voir se former des clusters représentant les 7 classes.

```python
def kmeans_apply(embed_data):
  kmeans = KMeans(n_clusters = 7)
  kmeans.fit(embed_data)
  labels = kmeans.labels_
  return labels
```

Maintenant que l'on a constitué les clusters, on peut les visualiser, ce qui nécessite une réduction de dimensions en utilisant PCA().

```python
def dim_reduction(vectors):
  # input = dataframe de l'espace vectoriel du vocabulaire.

  # Réduction dimension, PCA()
  pca = PCA(n_components = 2)
  matrix_pca = pca.fit_transform(vectors)

  # Mettre matrix_pca_tf dans un df
  df_pca = pd.DataFrame(matrix_pca, columns = ['pc_0', 'pc_1'])
  print("df_pca: {}".format(df_pca.shape))
  return df_pca

```

### Visualisation des clusters.

```python
def disp_res(vectors): # Affichage des résultats
  figure, axes = plt.subplots(1, 1, figsize = (5,5))
  print("appel fonction dim_reduction")
  print("sortiedim_reduction: {}".format(dim_reduction(vectors).shape))
  matrix = dim_reduction(vectors)
  print("matrix_shape: {}".format(matrix.shape))
  matrix['y_true'] = df['cat_lev_0_enc']
  matrix['y_pred'] = kmeans_apply(vectors) # vectors = embeddings
  print("matrix_shape_avt_affichage: {}".format(matrix.shape))


  sns.set_palette("bright")
  sns.scatterplot(data = matrix, x = 'pc_0', y = 'pc_1',
                  hue = matrix['y_true'],
                  palette = 'bright')

  plt.legend(loc='upper right')
  plt.title('représentation des documents en deux dimensions selon leur appartenance à une catégorie de niveau 0')
  plt.show()

  return matrix
```

### Text preprocessing avec bibliothèque Gensim.

```python
# Preprocess step (tokenize avec méthode package Gensim)

w2v_size=300
w2v_window=5
w2v_min_count=1
w2v_epochs=100
maxlen = 24 # adapt to length of sentences

sentences = df['product_name'].to_list()
print("avant transformation: {}".format(sentences[0:2]))


sentences = [gensim.utils.simple_preprocess(text, deacc=True, min_len=3, max_len=10) for text in sentences]
print("après transformation: {}".format(sentences[0:2]))
print("premier token: {}".format(sentences[0:2][0][0]))
```

### Création et entraînement du modèle Word2Vec

```python
# Création et entraînement du modèle Word2Vec

print("Build & train Word2Vec model ...")
w2v_model = gensim.models.Word2Vec(min_count=w2v_min_count, window=w2v_window,
                                                vector_size=w2v_size,
                                                seed=42,
                                                workers=1)

w2v_model.build_vocab(sentences)
print(w2v_model.build_vocab(sentences))


w2v_model.train(sentences, total_examples=w2v_model.corpus_count, epochs=w2v_epochs)
model_vectors = w2v_model.wv
print("vecteur correspondant au premier token enregistré dans 'index_to_key': {}".format(model_vectors[0]))
# print(model_vectors.vectors.shape)

w2v_words = model_vectors.index_to_key
print("premier et deuxième token stocké dans le dictionnaire 'index_to_key': {}".format(w2v_words[0:2]))

print("Vocabulary size: %i" % len(w2v_words))
print("Word2Vec trained")
```

### Construction des séquences.

```python
# Préparation des sentences (tokenization)

print("Fit Tokenizer ...")
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

# Construction matrice embedding
x_sentences = pad_sequences(tokenizer.texts_to_sequences(sentences),
                                                     maxlen=maxlen,
                                                     padding='post')
#print(sentences)
print(x_sentences)
print(len(x_sentences[0]))


num_words = len(tokenizer.word_index) + 1
print("Number of unique words: %i" % num_words)
```

### Construction matrice d'embedding.

```python
# Création de la matrice d'embedding

print("Create Embedding matrix ...")
w2v_size = 300

word_index = tokenizer.word_index
#print(word_index)

vocab_size = len(word_index) + 1
embedding_matrix = np.zeros((vocab_size, w2v_size))
i=0
j=0

for word, idx in word_index.items():
    i +=1
    if word in w2v_words:
        j +=1
        embedding_vector = model_vectors[word]
        if embedding_vector is not None:
            embedding_matrix[idx] = model_vectors[word]

word_rate = np.round(j/i,4)
print("Word embedding rate : ", word_rate)
print("Embedding matrix: %s" % str(embedding_matrix.shape))

#print(embedding_matrix)
```

### Construction correspondance séquences et matrice d'embedding.

```python
# Mettre en place une correspondance entre index des documents et tokens et leur représentation vectorielle
# dans embedding_matrix.

# Création embedding layer.
print("Création embedding layer...")

# 1. Taille vocabulaire ==> input_dim
input_dim = num_words
print("Nombre de mots uniques: {}".format(input_dim))

# 2. Dimension du vecteur de sortie associé au token ==> output_dim
output_dim = w2v_size
print("Dimension vecteur de sortie: {}".format(output_dim))

# 3. Taille des séquences ==> input_lenght
input_length = maxlen
print("Dimension des séquences d'entrée: {}".format(input_length))

# 4. Création embedding layer
embedding_layer = Embedding(input_dim = input_dim, output_dim = output_dim,
                            weights = [embedding_matrix],input_length = input_length)
print()
print("""Pour chaque index, on a créé une représentation vectorielle pour un mot donné.
En sortie de Embedding layer on a une matrice de taille nombre de séquences, nombre d'index
par séquence et nombre de features pour un index donné""")
print()
print("Embedding layer créée.")

print()
print("Réduction de dimension...")

# Calculer la valeur moyenne pour chaque vecteur
avg_embedding_model = GlobalAveragePooling1D()
avg_emb_val = avg_embedding_model(embedding_layer(x_sentences))
print(avg_emb_val)
print(avg_emb_val.shape)
print(type(avg_emb_val))
print()
print("""Fin de réduction de dimension. On a calculé la moyenne des vecteurs correspondant à une même séquence.
En sortie, on a une dimension en moins correspondant au nombre d'indices pour ue séquence""")




```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.manifold import TSNE
from nltk.tag.api import accuracy
```

```python
# Fonction qui permet de détecter les points singuliers de l'image.


def CalcFeatures(img):
  sift = cv2.SIFT_create()
  kp, des = sift.detectAndCompute(img, None)
  return des
```

```python
# Collecter tous les points singuliers (features) pour toutes les images.

features = []
for index, data in enumerate(l_images):
  if index <= nb_img:
    img_des = CalcFeatures(data)

    #print(img_des.shape)
    if img_des is not None:
      features.append(img_des)
      #print(len(features))
features_all = np.vstack((features))
print("Shape_features_all: {}".format(features_all.shape))
```

```python


# Calcul de la valeur de k = nb de clusters
#k = int(np.sqrt(len(features_all)))

k = 7 # k = 7 pour coïncider avec le nb de classes
print("Nb de clusters: {}".format(k))


# Constitution des clusters
kmeans = MiniBatchKMeans(n_clusters = k, init_size = 3*k, random_state = 42)
kmeans.fit(features_all)


# Tableaux aux memes dimensions
labels = kmeans.labels_
labels = labels.reshape(labels.shape[0], -1)

print("labels {}".format(labels[0:2]))
print("Shape_labels: {}".format(labels.shape))

labelled_descriptors = np.concatenate([features_all, labels], axis = 1)
print("Dimension matrice descripteurs pour l'ensemble des images: {}".format(labelled_descriptors.shape))



```

```python
# Histogramme des descripteurs par cluster
unique_labels = np.unique(labels)
print("unique_labels: {}".format(unique_labels))

l_counts = []
for label in unique_labels:
  count_descriptors = len(labelled_descriptors[labelled_descriptors == label])
  l_counts.append(count_descriptors)

df_counts_desc = pd.DataFrame(data = l_counts, columns = ['nb_desc'])
df_counts_desc['labels'] = unique_labels
print(df_counts_desc[0:2])
```
